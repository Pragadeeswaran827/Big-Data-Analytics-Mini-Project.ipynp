```python
import pandas as pd
df = pd.read_csv('01 renewable-share-energy.csv')
# or for Excel: pd.read_excel('renewable_energy.xlsx')
df.head()
```

```python
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

africa = df[df['Entity'] == 'Africa']
plt.plot(africa['Year'], africa['Renewables (% equivalent primary energy)'])
plt.xlabel('Year')
plt.ylabel('Renewables (% equivalent primary energy)')
plt.title('Renewable Energy Trend in Africa')
plt.show()
```

```python
grouped = df.groupby('Entity')['Renewables (% equivalent primary energy)'].mean()
print(grouped)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate mean renewable share for each region/entity
region_means = df.groupby('Entity')['Renewables (% equivalent primary energy)'].mean().sort_values(ascending=False)

# Plot top 7 regions/entities
top7 = region_means.head(7)
plt.figure(figsize=(10, 6))
sns.barplot(x=top7.values, y=top7.index, palette="viridis")
plt.xlabel('Mean Renewables (% equivalent primary energy)')
plt.title('Top 7 Regions by Average Renewable Energy Share')
plt.show()

# Display the values
print("Top 7 regions/entities by mean renewable share:")
print(top7)
```

**Conclusion**

This analysis helps visualize and compare renewable energy adoption across regions. We observe that some regions, like Vietnam and Western Africa, maintain higher average renewable shares, while the world average is much lower. Trends over time (such as Africa's recent increases) reflect improvement efforts in line with sustainable development goals (SDG 7: Affordable Energy). Further study can explore factors supporting rapid growth and predict future trends.
